```python
def multi_combination_lower_p0(combination_list):
    num_unique_events = len(combination_list)
    N = sum(combination_list)
    combination = int(''.join(map(str,combination_list)))
    counter = 0
    
    for i in range(combination+1):
        if (sum(list(map(int, str(i)))) == N):
            counter += 1
            print(number_extender(i, num_unique_events))
        
    return int(counter)
    

def number_extender(num, m):
    l = list()
    str_num = str(num)
    len_num = len(str_num)
    
    for i in range(m-len_num):
        l.append(0)
    for x in str_num:
        l.append(int(x))
    return l
        
    
```
