```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Einlesen des Datensatzes
covid = pd.read_csv("covid_de.csv")
```

**STEP 3.1:** Wo gab es die meisten Fälle in Deutschland? Listen Sie die Top 5 Bundesländer auf.

```python
# Wo gab es die meisten/wenigsten Fälle in Deutschland?
group_by_state_cases = covid.groupby('state').cases.sum().reset_index()
group_by_state_cases = group_by_state_cases.sort_values("cases", ascending=False).reset_index(drop=True)
group_by_state_cases.head()
```

**STEP 3.2:** Wo gab es die wenigsten Tode in Deutschland? Listen Sie die Top 5 Bundesländer auf.

```python
# Wo gab es die meisten/wenigsten Tote in Deutschland?
group_by_state_deaths = covid.groupby('state').deaths.sum().reset_index()
group_by_state_deaths = group_by_state_deaths.sort_values("deaths", ascending=False).reset_index(drop=True)
group_by_state_deaths.head()
```

**STEP 4:** Fügen Sie die Anzahl Fälle mit der Anzahl der Tode anhand der Bundesländer zusammen. Achten Sie darauf, dass keine Vermischungen von Zahlen von unterschiedlichen Bundesländern passieren.

```python
group_by_state_cases.set_index('state', inplace=True)
group_by_state_cases
```

```python
group_by_state_deaths.set_index('state', inplace=True)
group_by_state_deaths
```

```python
korrelation = group_by_state_cases
korrelation['deaths'] = group_by_state_deaths['deaths']
korrelation.head()
```

**STEP 5:** Stellen Sie einen einfach Bar-Plot da, welche die Bars von 'state' und 'deaths' anzeigt.

```python
# Gibt es eine Korrelation zwischen der Anzahl der Fälle und der Anzahl der Toten?
plt.figure(figsize=(25, 10))
plt.bar(korrelation.index.values, korrelation["deaths"])

plt.show()
```

```python
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
covid_month_list = [months[int(covid.date[i][5])*10 + int(covid.date[i][6])*1] for i in range(len(covid))]
covid_months = covid[["cases", "deaths", "recovered"]]
covid_months['month'] = covid_month_list
```

**STEP 6:** Gruppieren Sie die *covid_months* anhand der Monate, Summieren sie die Fälle auf und Sortieren Sie die Liste absteigend anhand der Anzahl der Fälle

```python
covid_months = covid_months.groupby('month').sum().sort_values('cases', ascending=False).reset_index()
covid_months
```

**STEP 7:** Fügen Sie zwei Spalten dem *covid_months* DataFrame hinzu.

* recovery_rate: Prozentualer Anteil von allen *recovered* Fällen
* death_rate: Prozentualer Anteil von allen *deaths* Fällen

Geben Sie das DataFrame am Ende aus und fragen Sie den Tutor, ob Sie alles richtig gemacht haben.

```python
covid_months['recovery_rate'] = covid_months.apply(lambda row: float(row.recovered/row.cases) if row.cases != 0 else 0, axis=1)
covid_months['death_rate'] = covid_months.apply(lambda row: float(row.deaths/row.cases) if row.cases != 0 else 0, axis=1)
covid_months
```
