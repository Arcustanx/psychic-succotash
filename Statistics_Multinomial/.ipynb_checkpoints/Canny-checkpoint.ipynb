```python
import cv2
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
import numpy as np
from matplotlib import pyplot as plt

im = cv2.imread('masken.jpg')
im = cv2.cvtColor(im, cv2.COLOR_RGB2BGR)

img = cv2.imread('masken.jpg', 0)
gauss = cv2.GaussianBlur(img,(5,5),2)
sobelx = cv2.Sobel(gauss,cv2.CV_64F,1,0,ksize=1)
sobely = cv2.Sobel(gauss,cv2.CV_64F,0,1,ksize=1)
betrag = (sobelx**2 + sobely**2)**(0.5)


plt.subplots(3, 2,figsize=(20,15))

plt.subplot(3,2,1),plt.imshow(im)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,2),plt.imshow(img,cmap = 'gray')
plt.title('Original Gray'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,4),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,5),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,6),plt.imshow(betrag,cmap = 'gray')
plt.title('Sobel Betrag'), plt.xticks([]), plt.yticks([])
plt.subplot(3,2,3),plt.imshow(gauss, cmap = 'gray')
plt.title('Gauss Filter'), plt.xticks([]), plt.yticks([])

plt.figure(figsize=(20,20))
plt.show()
```

```python
hyst_min = widgets.IntSlider(0,
                              min=0,
                              max=500,
                              step=10,
                              description='Min',
                              disabled=False,
                              continuous_update=False,
                              orientation='horizontal',
                              readout=True,
                              readout_format='d')

hyst_max = widgets.IntSlider(0,
                              min=0,
                              max=1000,
                              step=10,
                              description='Max',
                              disabled=False,
                              continuous_update=False,
                              orientation='horizontal',
                              readout=True,
                              readout_format='d')

sigma = widgets.IntSlider(20,
                              min=1,
                              max=100,
                              step=1,
                              description='Sigma',
                              disabled=False,
                              continuous_update=False,
                              orientation='horizontal',
                              readout=True,
                              readout_format='d')

def canny(a, b, c):  
    img = cv2.imread('masken.jpg',0)
    img_gauss = cv2.GaussianBlur(img,(5,5),c/10)
    edges = cv2.Canny(img_gauss,a,b)
    plt.subplots(1, 2,figsize=(20,20))
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Originales Bild'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Kantenbild'), plt.xticks([]), plt.yticks([])
    plt.show()

widgets.interact(canny, a=hyst_min, b=hyst_max, c=sigma);
```
